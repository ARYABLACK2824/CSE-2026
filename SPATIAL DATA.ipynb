```python
import folium
mymap = folium.Map(location=[21.0404, 82.4509], zoom_start=5)

locations = {
    "Presidency university": (13.1018, 77.3210),
    "ISKCON temple": (13.0035, 77.3303),
    "Laalbagh lake": (12.5645, 77.3459),
    "PARLE BISCUITS FACTORY": (13.0256, 77.2936),
    "DL Track Peenya": (13.0212, 77.3141)
}

for location, coordinates in locations.items():
    folium.Marker(location=coordinates, popup=location).add_to(mymap)

mymap
```
