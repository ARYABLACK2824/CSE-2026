```python
import requests
Link="https://en.wikipedia.org/wiki/Forbes_India%27s_richest_Indian_by_year"
Link_text=requests.get(Link).text
print(Link_text)
```

```python
from bs4 import BeautifulSoup
Soup = BeautifulSoup(Link_text,'lxml')
print(Soup)
```

```python
print(Soup.prettify())
```

```python
print(Soup.title)
```
